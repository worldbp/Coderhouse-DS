# Variable Target: Overall_Rating

La variable objetivo es de tipo númerica, así que podemos utilizar un modelo de tipo Aprendizaje supervisado y algoritmos de Regresión.

Se explorarán en consiguiente el comportamiento de la variable con algunos analisis estadísticos.




```python
target_variable = df['overall_rating']

target_variable.describe().round(2)
```

```python
#Distribución de variable target
media = target_variable.mean()
mediana = target_variable.median()
primer_cuartil = target_variable.quantile(0.25)
tercer_cuartil = target_variable.quantile(0.75)

print(media)
plt.figure()
# Histograma
sns.histplot(data = df,
             x = 'overall_rating',
             bins = 5,
             alpha = 0.25,
             discrete= True,
             kde = True,
             kde_kws = dict(
                 bw_adjust=9
             ),
             stat='probability')
# Agrego linea de promedio
plt.axvline(x = media,
            color = 'red',
            linestyle = '-',
            label = 'Promedio',
            linewidth = 3)
# Agrego linea de mediana
plt.axvline(x = mediana,
            color = '#11B613',
            linestyle = '-',
            label = 'Mediana',
            linewidth = 3)
# Agrego linea de primer cuartil
plt.axvline(x = primer_cuartil,
            color = 'black',
            linestyle = '--',
            label = 'Primer Cuartil',
            linewidth = 2)
# Agrego linea de tercer cuartil
plt.axvline(x = tercer_cuartil,
            color = 'black',
            linestyle = '--',
            label = 'Tercer Cuartil',
            linewidth = 2)

plt.legend()

plt.title('Distribución de Overall_Rating')

```

## Analisis de  variable Target: Overall_Rating

La variable objetivo sigue una distribución asimetrica neativa (la mediana es mayor que el promedio), concentrando una porción importante en reseñas con valores altos, y se observan pocos valores bajos.

Para este tipo de distibuciones, es apropiado utilizar una técnica de normalización que permita preparar nuestra data para otro tipo de análisis.

Se preesentará a continuación la distribución con

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

targetVarNorm = target_variable.values.reshape(-1, 1)

min_max_scaler = MinMaxScaler()

print(targetVarNorm)
X_train_minmax = min_max_scaler.fit_transform(targetVarNorm)


scaler = StandardScaler()
zscores = scaler.fit_transform(targetVarNorm)

logRatings = np.log(targetVarNorm)

fig, axs = plt.subplots(ncols=3, sharey= True)

sns.histplot(data = X_train_minmax,

             alpha = 0.1,
             kde = True,
             bins=10,
             stat='probability',
             color = '#B76BCB',
             kde_kws = dict(
                 bw_adjust=9
             ),
             ax=axs[0]
)
axs[0].set_title('Max-Min')

sns.histplot(data = zscores,
             alpha = 0.1,
             kde = True,
             bins=10,
             stat='probability',
             color = '#B76BCB',
             kde_kws = dict(
                 bw_adjust=9
             ),
             ax=axs[1]
)
axs[1].set_title('Z-Scores')

sns.histplot(data = logRatings,


             alpha = 0.1,
             kde = True,
             bins=10,
             stat='probability',
             color = '#B76BCB',
             kde_kws = dict(
                 bw_adjust=9
             ),
             ax=axs[2]
)
axs[2].set_title('Log')

```

### 1.   Outlook, Recommend, CEO_Approv:
Para poder realizar nuestro análisis supervisado, vamos a tener que convertir algunas de nuestras variables cualitativas que tienen carácter ordinal en variables númericas para su procesamiento.
Las recomendaciones de la compañia , CEO , y futuro de la empresa le han colocado las categorías v, r, x, and o, con el siguiente significado:
v - Positivo, r - Más o Menos, x - Negativo, o - Sin opinión

A continuación exploraremos esas variables para entender su comportamiento:

```python
df_ordinales = df[['recommend', 'ceo_approv', 'outlook']]
print(
    df_ordinales['recommend'].value_counts()/len(df_ordinales)*100, '\n\n',

    df_ordinales['ceo_approv'].value_counts()/len(df_ordinales)*100, '\n\n',

    df_ordinales['outlook'].value_counts()/len(df_ordinales)*100

)
```

```python
import random

#No estamos seguros de que significa cuando una persona dice Más o Menos, entonces generamos un numero al azar entre ambos extremos.
#Otra solución posible sería hacer la escala del 1-3 para que no hay espacio para la interpretación.

dict_ordinal_cat = {'v': 5, 'x': 1, 'r': random.randint(2,4), 'o': np.NaN }

for col in df_ordinales.columns:
    print(col)
    df_ordinales[f'{col}']= df_ordinales[f'{col}'].applymap(dict_ordinal_cat)


print(
    df_ordinales['recommend'].value_counts()/len(df_ordinales)*100, '\n\n',

    df_ordinales['ceo_approv'].value_counts()/len(df_ordinales)*100, '\n\n',

    df_ordinales['outlook'].value_counts()/len(df_ordinales)*100

)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()



sentiment_scores = df[['headline', 'cons', 'pros']].applymap(lambda x: analyzer.polarity_scores(str(x)), na_action='ignore')
print(sentiment_scores['headline'][286369	] )

```

```python
# sentiment_scores.rename( columns={'headline':'headline_num', 'cons':'cons_num', 'pros':'pros_num'})

sentiment_scores = sentiment_scores.applymap(lambda x: x['compound'], na_action='ignore' )
```

 ##### Variance Inflation Factor



El VIF mide cuánto aumenta la varianza (o incertidumbre) de un coeficiente de regresión estimado cuando tus predictores (variables) están correlacionados.

Los valores altos de VIF indican alta correlación entre un predictor y otros predictores, sugiriendo que la información de este predictor es en parte redundante.

Los valores bajos de VIF (generalmente por debajo de 5 o 10, dependiendo del contexto) significan que un predictor no está altamente correlacionado con otros predictores, facilitando la interpretación de su impacto en la variable de resultado.

Interpretación de los valores de VIF:

*    VIF = 1: No hay correlación entre el predictor y otras variables. No hay problemas con la multicolinealidad.

*    VIF entre 1 y 5: Correlación moderada. En la mayoría de los casos, no causa problemas significativos.*

*    VIF por encima de 5: Alta correlación. El predictor está altamente correlacionado con otras variables en el modelo, lo que dificulta interpretar su efecto individual.*

Magnitud de los Puntajes de Carga:

Puntajes de carga positivos grandes (cerca de 1) indican una fuerte correlación positiva entre la característica original y el componente principal. Un aumento en el valor de la característica original corresponde a un aumento en el componente principal.
Puntajes de carga negativos grandes (cerca de -1) indican una fuerte correlación negativa. Un aumento en el valor de la característica original corresponde a una disminución en el componente principal.

Signo de los Puntajes de Carga:

El signo del puntaje de carga indica la dirección de la correlación.
Puntajes de carga positivos significan que la característica original y el componente principal aumentan o disminuyen juntos.
Puntajes de carga negativos significan que cambian en direcciones opuestas.
Comparación de Magnitudes:

Comparar la magnitud de los puntajes de carga entre características para un solo componente principal puede revelar qué características originales tienen una mayor influencia en ese componente.
Las características con puntajes de carga absolutos más altos contribuyen más a la variación explicada por el componente principal.
Puntajes de Carga Cercanos a Cero:

Un puntaje de carga cercano a cero (alrededor de 0) indica una correlación débil o nula entre la característica original y el componente principal. Esta característica tiene poca influencia en ese componente principal en particular.
Múltiples Componentes Principales:

Cada componente principal representa un patrón diferente de variación en los datos. La interpretación se vuelve más significativa al observar varios componentes principales juntos.
Las características con puntajes de carga significativos en varios componentes principales pueden tener relaciones complejas con los patrones subyacentes en los datos.
Varianza Explicada:

Considera la proporción de varianza explicada de cada componente principal. Una varianza explicada más alta indica que el componente principal captura una mayor proporción de la varianza total en el conjunto de datos.
Los componentes principales con una mayor proporción de varianza explicada son más informativos y tienen un mayor impacto en la explicación de la variabilidad del conjunto de datos.

**Error Absoluto Medio (MAE):**

MAE mide el promedio de los errores absolutos entre los valores predichos y los valores reales.
Un MAE más bajo indica una mayor precisión. En tu caso, un MAE de 0.46-0.5 sugiere que, en promedio, las predicciones tienen un error de 0.46-0.5 unidades respecto a los valores reales. Considerando que el rango de nuestro valores discretos y cardinalidad este valor es muy alto para que el modelo pueda ser de ayuda.

**Error Cuadrático Medio (MSE):**

MSE mide el promedio de las diferencias al cuadrado entre los valores predichos y los valores reales.
Al igual que el MAE, valores de MSE más bajos indican una mayor precisión. Tu MSE de 0.4 significa que los errores al cuadrado están, en promedio, a 0.4 unidades de los valores reales.

**Coeficiente de Determinación (R-cuadrado):**

R-cuadrado representa la proporción de la varianza en la variable dependiente (objetivo) que es predecible a partir de las variables independientes (características).
Los valores de R-cuadrado van de 0 a 1. Un R-cuadrado más alto (más cercano a 1) indica un mejor ajuste del modelo a los datos. Un R-cuadrado de 0.96 sugiere que aproximadamente el 71% de la variabilidad en la variable objetivo es predecible a partir de las características en tu modelo.

**Interpretación:**

El modelo tiene un MAE y MSE no despreciables, lo que indica que hace predicciones tienen errores no  pequeños en promedio.

El  valor de R-cuadrado (0.71 y 0.66) indica que el modelo explica un 70% como mucho de la variabilidad en la variable objetivo, lo que sugiere que necesitamos ajsutar nuestro modelo para ayudarnos a mejorarlo.
